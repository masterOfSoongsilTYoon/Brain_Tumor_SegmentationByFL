```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
efficient_deeplab = pd.read_csv("./CSV/Result_Brain_Baseline_deeplabV3_Backbone_efficient.csv")
res101_deeplab = pd.read_csv("./CSV/Result_Brain_Baseline_deeplabV3_Backbone_res101.csv")
unet = pd.read_csv('./CSV/Result_Brain_Baseline_Unet.csv')
```

```python
plt.plot(efficient_deeplab["loss"], color="g", label="efficient_deeplab")
plt.plot(res101_deeplab["loss"], color="b", label="res101_deeplab")
plt.plot(unet["loss"], color="r", label="Unet")
plt.legend()
plt.xlabel("epochs")
plt.title("Loss By Each Model")
plt.show()
```

```python
plt.plot(efficient_deeplab["mIOU"], color="g", label="efficient_deeplab")
plt.plot(res101_deeplab["mIOU"], color="b", label="res101_deeplab")
plt.plot(unet["mIOU"], color="r", label="Unet")
plt.legend()
plt.xlabel("epochs")
plt.title("mIOU By Each Model")
plt.show()
```
